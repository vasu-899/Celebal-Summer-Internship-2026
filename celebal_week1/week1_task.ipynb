```python
import pandas as pd
df = pd.read_csv('Combined_dataset.csv')
```

```python
print(df.head())
print(df.shape)
print(df.columns)
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
df['what_customers_said'] = df['what_customers_said'].fillna('No reviews')
df['seller_name'] = df['seller_name'].fillna('Unknown Seller')
df['seller_information'] = df['seller_information'].fillna('Not Available')
df['variations'] = df['variations'].fillna('No variations')
df['videos'] = df['videos'].fillna('No videos')
df['discount'] = df['discount'].fillna(0)

print(df.isnull().sum())
```

```python

print(df[['title', 'initial_price', 'discount', 'rating']].head())

filtered = df[df['rating'] > 4.0]
print("\nHigh rated products:", filtered.shape)
```

```python
print("Before:", df.shape)
df.drop_duplicates(inplace=True)
print("After:", df.shape)
```

```python
df['total_amount'] = df['initial_price'] * (1 - df['discount'] / 100)
print(df[['title', 'initial_price', 'discount', 'total_amount']].head())
```

```python
df.to_csv('cleaned_shopping_data.csv', index=False)
print("File saved!")
```

## Summary
- Dataset loaded: 1000 rows, 24 columns
- Missing values found in: discount, what_customers_said, seller_name, seller_information, variations, videos -> filled with appropriate default values
- Duplicates removed: 0 duplicates found
- New column added: total_amount = initial_price * (1 - discount/100)
- Cleaned file saved as cleaned_shopping_data.csv
